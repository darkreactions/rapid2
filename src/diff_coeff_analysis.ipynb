```python
import pandas as pd
import numpy as np
from pathlib import Path
from collections import defaultdict
```

```python
dcm_dmso_c = pd.read_csv('laser_diffusion_coefficient_measurement_DCM_DMSO_L50_c.csv')
dcm_gbldmf_b = pd.read_csv('laser_diffusion_coefficient_measurement_DCM_GBLDMF_L_50_b.csv')
dcm_gbldmf_a = pd.read_csv('laser_diffusion_coefficient_measurement_DCM_GBLDMF_L_50_a.csv')
dcm_dmf_c = pd.read_csv('laser_diffusion_coefficient_measurement_DCM_DMF_L50_c.csv')
```

```python
path = Path('./')
pathlist = list(path.glob(pattern='laser_diffusion_coefficient_measurement_*'))
dfs = defaultdict(dict)
for path in pathlist:
    split_name = path.name.split('_')
    solvent = split_name[5]
    set_name = split_name[7][0]
    dfs[set_name][solvent] = pd.read_csv(path)
```

```python
def normalize_dist(df):
    return (df['distance_cm']-df['distance_cm'].min())/(df['distance_cm'].max() - df['distance_cm'].min())
```

```python
for set_name in dfs:
    for solvent in dfs[set_name]:
        df = dfs[set_name][solvent]
        df['normed_dist'] = normalize_dist(df)
```

```python
import plotly.graph_objects as go
data = []
set_name = 'c'
for solvent in dfs[set_name]:
    df = dfs[set_name][solvent]
    data.append(go.Scatter(x=df['elapsed hours'], y=df['normed_dist'], name=f'{solvent}_{set_name}'))
fig = go.Figure(data=data)
fig.update_layout(
    title=f"Set Name {set_name}",
    xaxis_title="Elapsed Time",
    yaxis_title="Normalized Distance",)
fig.show()
```

```python
import plotly.graph_objects as go

data = [go.Scatter(x=dcm_dmso_c['elapsed hours'], y=dcm_dmso_c['normed_dist'], name='DCM_DMSO_c'),
        go.Scatter(x=dcm_dmf_c['elapsed hours'], y=dcm_dmf_c['normed_dist'], name='DCM_DMF_c'),
        go.Scatter(x=dcm_gbldmf_b['elapsed hours'], y=dcm_gbldmf_b['normed_dist'], name='DCM_GBLDMF_b'),
        go.Scatter(x=dcm_gbldmf_a['elapsed hours'], y=dcm_gbldmf_a['normed_dist'], name='DCM_GBLDMF_a'),]
fig = go.Figure(data=data)
fig.show()
```
